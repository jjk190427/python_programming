```python
#필요한 패키지 불러오기
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
```

```python
# 파일 경로 지정
p_library_20 = "성동구립성수도서관 장서 대출목록 (2020년 11월).csv"
p_library_23 = "성동구립성수도서관 장서 대출목록 (2023년 11월).csv"

# CSV 파일을 DataFrame으로 읽어오기 (인코딩은 'cp949'를 사용)
d_library_20 = pd.read_csv(p_library_20, encoding='cp949')
d_library_23 = pd.read_csv(p_library_23, encoding='cp949')
```

```python
# 특수문자를 제거하는 정규표현식을 활용하여 데이터프레임의 특정 열에서 특수문자 제거

# 2020년 데이터프레임에서 특수문자 제거
d_library_20['도서명'] = d_library_20['도서명'].str.replace(r"[^\w]", "")
d_library_20['저자'] = d_library_20['저자'].str.replace(r"[^\w]", "")
d_library_20['출판사'] = d_library_20['출판사'].str.replace(r"[^\w]", "")
d_library_20['발행년도'] = d_library_20['발행년도'].str.replace(r"[^\w]", "")

# 2023년 데이터프레임에서 특수문자 제거
d_library_23['도서명'] = d_library_23['도서명'].str.replace(r"[^\w]", "")
d_library_23['저자'] = d_library_23['저자'].str.replace(r"[^\w]", "")
d_library_23['출판사'] = d_library_23['출판사'].str.replace(r"[^\w]", "")
d_library_23['발행년도'] = d_library_23['발행년도'].str.replace(r"[^\w]", "")
```

3. 2020년 11월과 2023년 11월 가장 대출이 많이 발생한 도서 상위 20권의 리스트와 그 빈도수 비교

```python
# 2020년과 2023년 각각 중복되지 않은 ISBN을 기준으로 대출건수를 합산하고 대출건수 기준으로 정렬
dup_d_library_20 = d_library_20.drop_duplicates(['ISBN'])
dup_d_library_23 = d_library_23.drop_duplicates(['ISBN'])

#ISBN 그룹으로 묶어 대출건수를 합해서 도서의 총 대출건수를 계산
d_library_20_sum = d_library_20.groupby('ISBN').sum()['대출건수'].sort_values(ascending=False)
d_library_23_sum = d_library_23.groupby('ISBN').sum()['대출건수'].sort_values(ascending=False)

#중복제거된 데이터프레임에 합산된 대출건수 추가
dup_d_library_20_sum = pd.merge(dup_d_library_20, d_library_20_sum, on='ISBN')
dup_d_library_23_sum = pd.merge(dup_d_library_23, d_library_23_sum, on='ISBN')

#대출건수를 기준으로 내림차순 정렬
dup_d_library_20_sum = dup_d_library_20_sum.sort_values(by='대출건수_y',ascending=False)
dup_d_library_23_sum = dup_d_library_23_sum.sort_values(by='대출건수_y',ascending=False)

#필요없는 컬럼 제거
dup_d_library_20_sum = dup_d_library_20_sum.drop(['대출건수_x','번호','저자','출판사','세트 ISBN','부가기호','권','주제분류번호','도서권수','등록일자'], axis=1)
dup_d_library_23_sum = dup_d_library_23_sum.drop(['대출건수_x','번호','저자','출판사','세트 ISBN','부가기호','권','주제분류번호','도서권수','등록일자'], axis=1)

#명확한 데이터 조회를 위해 컬럼명 변경
dup_d_library_20_sum = dup_d_library_20_sum.rename(columns={'도서명':'2020년 도서명'})
dup_d_library_20_sum = dup_d_library_20_sum.rename(columns={'발행년도':'2020년 발행년도'})
dup_d_library_20_sum = dup_d_library_20_sum.rename(columns={'대출건수_y':'2020년 대출건수'})

dup_d_library_23_sum = dup_d_library_23_sum.rename(columns={'도서명':'2023년 도서명'})
dup_d_library_23_sum = dup_d_library_23_sum.rename(columns={'발행년도':'2023년 발행년도'})
dup_d_library_23_sum = dup_d_library_23_sum.rename(columns={'대출건수_y':'2023년 대출건수'})

#대출건수가 많은 순서대로 상위 20개 서열을 도출하여 데이터프레임에 추가
dup_d_library_20_sum = dup_d_library_20_sum.head(20)
dup_d_library_23_sum = dup_d_library_23_sum.head(20)

#데이터에서 대출건수 많은 순서대로 서열 지정후 데이터프레임에 추가
dup_d_library_20_sum['서열'] = dup_d_library_20_sum['2020년 대출건수'].rank(ascending=False)
dup_d_library_23_sum['서열'] = dup_d_library_23_sum['2023년 대출건수'].rank(ascending=False)

#서열을 인덱스로 지정
dup_d_library_20_sum.set_index(['서열'], inplace=True)
dup_d_library_23_sum.set_index(['서열'], inplace=True)

# 두 데이터프레임 결합
compare_dl = pd.concat([dup_d_library_20_sum, dup_d_library_23_sum])

# 인덱스 초기화 및 결과 출력
compare_dl.reset_index(drop = True)
compare_dl
```

5. 2020년 11월 가장 대출이 많이 발생한 도서 상위 50권의 당시 대출 횟수와 현재 대출 횟수 비교하여 증가한 도서와 감소한 도서 파악, 그래프로 표현

```python
# 2020년과 2023년 각각 중복되지 않은 ISBN을 기준으로 대출건수를 합산하고 대출건수 기준으로 정렬
dup_d_library_20 = d_library_20.drop_duplicates(['ISBN'])
dup_d_library_23 = d_library_23.drop_duplicates(['ISBN'])

#ISBN 그룹으로 묶어 대출건수를 합해서 도서의 총 대출건수를 계산
d_library_20_sum = d_library_20.groupby('ISBN').sum()['대출건수'].reset_index().sort_values(by='대출건수', ascending=False)
d_library_23_sum = d_library_23.groupby('ISBN').sum()['대출건수'].reset_index().sort_values(by='대출건수', ascending=False)

#중복제거된 데이터프레임에 합산된 대출건수 추가
dup_d_library_20_sum = pd.merge(dup_d_library_20, d_library_20_sum, on='ISBN')
dup_d_library_23_sum = pd.merge(dup_d_library_23, d_library_23_sum, on='ISBN')

#대출건수를 기준으로 내림차순 정렬
dup_d_library_20_sum = dup_d_library_20_sum.sort_values(by='대출건수_y', ascending=False)
dup_d_library_23_sum = dup_d_library_23_sum.sort_values(by='대출건수_y', ascending=False)

#필요없는 컬럼 제거
dup_d_library_20_sum = dup_d_library_20_sum.rename(columns={'도서명':'2020년 도서명', '발행년도':'2020년 발행년도', '대출건수_y':'2020년 대출건수'})
dup_d_library_23_sum = dup_d_library_23_sum.rename(columns={'도서명':'2023년 도서명', '발행년도':'2023년 발행년도', '대출건수_y':'2023년 대출건수'})


# ISBN을 기준으로 20년과 23년의 도서 데이터를 내부 조인하여 병합
merge_lb2 = pd.merge(dup_d_library_20_sum, dup_d_library_23_sum, on='ISBN', how='inner')

# 대출건수의 변화를 나타내는 새로운 컬럼 'change_count'를 추가
merge_lb2['change_count'] = merge_lb2['2020년 대출건수'] - merge_lb2['2023년 대출건수']

# 결과 출력
merge_lb2.head(50)
```

```python
# 그래프 생성
fig, ax = plt.subplots()

# 점 그래프 생성
scatter = ax.scatter(merge_lb2['change_count'], merge_lb2['2020년 대출건수'])

# 각 점에 도서명 표시
for i, txt in enumerate(merge_lb2['2020년 도서명']):
    ax.annotate(txt, (merge_lb2['change_count'][i], merge_lb2['2020년 대출건수'][i]))

# 그래프 제목 및 레이블 설정
plt.title('대출건수 변화에 따른 도서명')
plt.xlabel('대출건수 변화')
plt.ylabel('2020년 대출건수')

# 그래프 표시
plt.show()
```
